举例1：从物理磁盘的yaml文件中读取prompt数据  ,这个例子讲的是 Prompt 的持久化（Prompt Persistence）

```python
from langchain_core.prompts import load_prompt
import dotenv

dotenv.load_dotenv()

prompt = load_prompt("prompt.yaml", encoding="utf-8")
# print(prompt)
#print(prompt.format(name="年轻人", what="滑稽"))
print("===== Prompt对象 =====")
print(prompt)

print("\n===== Prompt类型 =====")
print(type(prompt))

print("\n===== 模板内容 =====")
print(prompt.template)

print("\n===== 输入变量 =====")
print(prompt.input_variables)

print("\n===== 格式化结果 =====")
print(
    prompt.format(
        name="年轻人",
        what="滑稽"
    )
)
```
